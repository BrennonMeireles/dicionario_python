## Crie uma lista locadora de contenha os seguintes elementos:
Título = Star Wars
Ano = 1977
Diretor = George Lucas


---


Título = Superman
Ano = 1978
Diretor = Richard Donner


---


Título = Matrix
Ano = 1999
Diretor = Wachowsky

```python
#23/09/2023


film1 = {'Título':'Star Wars','Ano':'1977','Diretor':'George Lucas'} # dicionario do filme Star wars com elementos de ano e diretor
film2 = {'Título':'Superman','Ano':'1978','Diretor':'Richard Donner'} # dicionario do filme Superman
film3 = {'Título':'Matrix','Ano':'1999','Diretor':'Wachowsky'} # dicionario do filme Matrix

locadora = list() # criação da lista vazia

locadora.append(film1) # adição do dicionario para dentro da lista
locadora.append(film2) # .
locadora.append(film3) # .

print(locadora) # exibição da lista locadora com todos os dicionarios...

for i in locadora:
  print(f'Filme: {i["Título"]} \nAno: {i["Ano"]} \nDireção: {i["Diretor"]}\n')
```

```python
# Outro jeito de criar um dicionario...

tabela = {} # criação de um dicionario

tabela['Alface'] = 0.45 # adição de cada termo e valor
tabela['Batata'] = 1.20 # .
tabela['Tomate'] = 2.30 # .
tabela['Feijão'] = 1.50 # .

print(tabela) # exibir a lista sem organização...

for prod,preco in tabela.items(): # for para separar os itens e exibilos separadamente
  print(f'{prod} valor: R$ {preco:3.2f}') # 3.2f para colocar numeros com duas casaz decimais
```
